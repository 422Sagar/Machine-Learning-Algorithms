```python
import pandas as pd
import numpy as np
```

```python
# Reading the Diabetes Data 
Diabetes = pd.read_csv("C:\\ASSIGNMENTS\\Decision Tree and Random Forest\\Diabetes_RF.csv")
```

```python
Diabetes.head(5)
```

```python
Diabetes.columns
```

```python
colnames = list(Diabetes.columns)
```

```python
predictors = colnames[:8]
target = colnames[8]
```

```python
# Storing thr predictor and target values in X and Y variables
X = Diabetes[predictors]
Y = Diabetes[target]
```

# RF model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_jobs=-1,oob_score=True,n_estimators=1000,criterion="entropy")

# n_estimators -> Number of trees ( you can increase for better accuracy)
# n_jobs -> Parallelization of the computing and signifies the number of jobs running parallel for both fit and predict
# oob_score = True means model has done out of box sampling to make predictions
```

```python
np.shape(Diabetes)
```

```python
# Fitting RandomForestClassifier model from sklearn.ensemble 
rf.fit(X,Y) 
```

```python
# Attributes that comes along with RandomForest function
rf.estimators_
```

```python
rf.classes_ # class labels (output)
```

```python
rf.n_classes_ # Number of levels in class labels 
```

```python
rf.n_features_  # Number of input features in model 8 here.
```

```python
rf.n_outputs_ # Number of outputs when fit performed
```

```python
rf.oob_score_ # 0.72916
```

```python
Diabetes['rf_pred']=rf.predict(X)
```

```python
cols = ['rf_pred',' Class variable']
```
